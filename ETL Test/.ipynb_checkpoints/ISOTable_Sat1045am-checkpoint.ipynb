```python
#ensure Chrome Driver is in the same directory as this notebook.   (Different for MAC user)
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)

#navigate to the site which will be scraped
url = 'https://www.nationsonline.org/oneworld/country_code_list.htm'
browser.visit(url)
```

```python
# Retrieve page with the requests module
response = requests.get(url)

# Create BeautifulSoup object of this URL with the featured image; parse with 'html.parser'
soup = bs(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
# results for only the table
apple = soup.find('table', id='CountryCode')
```

```python
print(apple.prettify())
```

```python
# Print all rows WORKING
peach = apple.find_all('td')
for row in peach:
    print('--------------')
    print(row.text.strip())
```

```python
#first, determine the data in the rows.  Need the index number
peach = apple.find_all('tr')
for row in peach:
    obtainColumns = row.find_all('td')
    if (len(obtainColumns) == 5):
        #print(country)
        print("++++++++++++ NEW ROW ++++++++++++++++")
        print("HAS LENGTH OF " + str(len(obtainColumns)))
    for col in obtainColumns:
        print('--------------')
        print()
        print(col.text.strip())
```

```python
# Print all rows WORKING
countrylist = []
iso2list = []
iso3list = []
unlist = []

peach = apple.find_all('tr')
for row in peach:
    obtainColumns = row.find_all('td')
    if (len(obtainColumns) == 5):
        
        #find the countries
        country = obtainColumns[1].text
        countrylist.append(country)
        
        #find the 2 character ISO Code
        iso2 = obtainColumns[2].text
        iso2list.append(iso2)
     
        #find the 3 character ISO Code
        iso3 = obtainColumns[3].text
        iso3list.append(iso3)
    
        #find the UN code
        un = obtainColumns[4].text
        unlist.append(un)
    
        #print(country)
    #print("++++++++++++ NEW ROW ++++++++++++++++")
    #print("HAS LENGTH OF " + str(len(obtainColumns)))
    # for col in obtainColumns:
    #    print('--------------')
    #    print()
    #    print(col.text.strip())
print(countrylist)
```
