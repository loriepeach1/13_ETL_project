```python
#ensure Chrome Driver is in the same directory as this notebook.   (Different for MAC user)
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)

#navigate to the site which will be scraped
url = 'https://www.nationsonline.org/oneworld/country_code_list.htm'
browser.visit(url)
```

```python
# Retrieve page with the requests module
response = requests.get(url)

# Create BeautifulSoup object of this URL with the featured image; parse with 'html.parser'
soup = bs(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
# results for only the table
table = soup.find('table', id='CountryCode')
```

```python
print(table.prettify())
```

```python
# Print all rows
rows = table.find_all('td')
for row in rows:
    print('--------------')
    print(row.text.strip())
```

```python
#determine the data in the rows.  Need the index number for the individual values
rows2 = table.find_all('tr')
for row in rows2:
    obtainColumns = row.find_all('td')
    if (len(obtainColumns) == 5):
        #print(country)
        print("++++++++++++ NEW ROW ++++++++++++++++")
        print("HAS LENGTH OF " + str(len(obtainColumns)))
    for col in obtainColumns:
        print('--------------')
        print()
        print(col.text.strip())
```

```python
# Print all rows WORKING
countrylist = []
iso2list = []
iso3list = []
unlist = []

# rows2 = table.find_all('tr')
for row in rows2:
    obtainColumns = row.find_all('td')
    if (len(obtainColumns) == 5):
        
        #find the countries
        country = obtainColumns[1].text.strip()
        countrylist.append(country)
        
        #find the 2 character ISO Code
        iso2 = obtainColumns[2].text.strip()
        iso2list.append(iso2)
     
        #find the 3 character ISO Code
        iso3 = obtainColumns[3].text.strip()
        iso3list.append(iso3)
    
        #find the UN code
        un = obtainColumns[4].text.strip()
        unlist.append(un)
    
        #print(country)
    #print("++++++++++++ NEW ROW ++++++++++++++++")
    #print("HAS LENGTH OF " + str(len(obtainColumns)))
    # for col in obtainColumns:
    #    print('--------------')
    #    print()
    #    print(col.text.strip())
```

```python
# Close the browser after scraping
browser.quit()
```

```python
# now that there are 4 lists, create a panda dataframe
scrapeISO_df = pd.DataFrame(list(zip(countrylist, iso2list, iso3list, unlist)),
              columns=['country','2_char_ISO_Code', '3_char_ISO_Code', 'UN_Code'])
```

```python
scrapeISO_df.head(20)
```

```python
#convert the dataframe into a csv

scrapeISO_df.to_csv('ISO_Clean.csv', encoding="ISO-8859-1")

#the csv file can now be used to create a sql table
```
